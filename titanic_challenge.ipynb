```python
import pandas as pd
import warnings

df=pd.read_csv("train.csv")
test=pd.read_csv('test.csv')

test_id=test['PassengerId']

warnings.filterwarnings("ignore")
```

```python
def clean(data):
    data=data.drop(["PassengerId","Ticket","Cabin","Name","Fare"],axis=1)

    data["Age"].fillna(data['Age'].median(),inplace=True)
    
    return data

df=clean(df)
test=clean(test)
```

```python
df.head(5)
```

```python
def numeric(data):
    data=pd.get_dummies(data=data,columns=["Sex","Embarked"],drop_first=True,dtype=int)

    return data

df=numeric(df)
test=numeric(test)
```

```python
df.head(5)
```

```python
from sklearn.model_selection import train_test_split

y=df['Survived']
x=df.drop(['Survived'],axis=1)

x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=1,test_size=0.80)
```

```python
from sklearn.linear_model import LogisticRegression

regressor=LogisticRegression(random_state=1,max_iter=1000)
regressor.fit(x_train,y_train)

y_pred=regressor.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test,y_pred)
```
